### <span style="color:red">1. Which city had the most amount of accidents ?</span>

```python
# create a dataframe of city and their corresponding accident cases
city_df = pd.DataFrame(df_cleaned['City'].value_counts()).reset_index().rename(columns={'index':'City', 'City':'Cases'})
city_df
```

```python
city_df["Perc_Cases"] = (city_df.Cases/city_df.Cases.sum())*100
city_df
```

```python
top_10_cities = city_df.head(10)
top_10_cities.set_index("City", inplace=True)
plt.figure(figsize=((15,8)))
graph = plt.bar(top_10_cities.index,top_10_cities.Cases)
plt.title("Top 10 Cities in US with most no. of Road Accident Cases")
plt.ylabel("Accident Cases")
plt.xlabel("Cities")
plt.xticks(rotation=10, fontsize=12)
plt.yticks(fontsize=12)

i = 0
for p in graph:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    plt.text(x+width/2,
             y+height*1.01,
             str(top_10_cities.Perc_Cases[i])[0:4]+'%',
             ha='center',
             )
    i+=1
plt.show()
```

### <span style="color:green"> 1. Houston is the city with highest (2.33%) no. of road accidents in US .<br><br>2. Charlotte is the city with 2nd highest (2.08%) no. of road accidents in US .</span>

### <span style="color:red">2. At what time of the year or day that had more reported accidents ? Infer why ?</span>

```python
df_cleaned["_Count"] = 1
```

```python
df_cleaned.groupby(pd.Grouper(key='_Weather_Timestamp_Date', axis=0, freq='Y')).sum()[["_Count"]]
```

#### We can see data is not balanced per year. Reasons can be
- There is increase in no of accidents per year as the number says
- There is something missing while collecting data

```python
df_by_year = df_cleaned.set_index("_Weather_Timestamp_Date")
df_by_year.head()
```

```python
df_by_year.groupby([(df_by_year.index.year),(df_by_year.index.month)])[["_Count"]].sum()
```

```python
temp = df_by_year.groupby([(df_by_year.index.year),(df_by_year.index.month)])[["_Count"]].sum()

temp.plot(figsize=((15,8)), style="-o")

plt.title("Overall montly no. of Road Accident Cases")
plt.ylabel("Accident Cases")
plt.xlabel("Date")
plt.xticks(rotation=10, fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
df_by_year.groupby(df_by_year.index.month)[["_Count"]].sum()
```

```python
temp = df_by_year.groupby(df_by_year.index.month)[["_Count"]].sum()

temp.plot.bar(figsize=((15,8)))

plt.title("Over 5 years Montly no. of Road Accident Cases")
plt.ylabel("Accident Cases")
plt.xlabel("Month")
plt.xticks(rotation=10, fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

### <span style="color:green">From this we can clearly state that Most of the accidents occurs in the months of sept to December <br><br> Reason behind this can be the winter season where fog may affect the visibility of raod causing accidents. Or more people travel during this season for holidays , vacations or trips</span>
